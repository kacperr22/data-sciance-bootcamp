```
fig = plt.figure(figsize=(12, 10))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('children.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
```

```
fig, ((p1, p2), (p3, p4)) = plt.subplots(2,2)
p1.imshow(mpimg.imread('office.jpg'))
plt.axis('Off')
p2.imshow(mpimg.imread('business.jpg'))
plt.axis('Off')
p3.imshow(mpimg.imread('children.jpg'))
plt.axis('Off')
p4.imshow(mpimg.imread('workplace.jpg'))
plt.axis('Off')
```

```
fig = plt.figure(figsize=(12, 10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(220 + idx + 1)
    plt.imshow(mpimg.imread(img))
```

```
fig = plt.figure(figsize=(12, 10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(140 + idx + 1)
    plt.imshow(mpimg.imread(img))
```
