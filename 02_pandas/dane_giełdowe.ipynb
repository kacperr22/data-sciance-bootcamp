```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetches stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('UBER')
df.info()
```

```python
df = df.head(150)
```

```python
df
```

```python
df.Average = (df.High + df.Low) /2 ## albo df['Average'] = (df.High + df.Low) /2
```

```python
df
```

```python
df1 = df.sort_index()
df1
```

```python
df1['Close_shift'] = df.Close.shift(1) -1

df1
```

```python
df1['up/down[%]'] = round((df1.Close - df1.Close_shift ) / df1.Close * 100, 3)
df1
```
